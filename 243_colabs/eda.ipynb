## processing data

```python
def plot_length_energy(dictionary, model_name, color='c', showplot=True):
  model_data = dictionary[model_name]['model']

  plt.scatter(model_data["response_length"], model_data["energy"], alpha=0.2, s=5, color=color)
  plt.legend([model_name], loc='upper left')

  if showplot:
    plt.show()
```

```python
def plot_length_latency(dictionary, model_name, color='c', showplot=True):
  model_data = dictionary[model_name]['model']

  plt.scatter(model_data["response_length"], model_data["latency"], alpha=0.2, s=5, color=color)
  plt.legend([model_name], loc='upper left')

  if showplot:
    plt.show()
```

```python
def fit_energy(dictionary, model_name):
  model_data = dictionary[model_name]['model']

  X = model_data["response_length"]
  y = model_data["energy"]

  #X = sm.add_constant(X)

  model = sm.OLS(y, X).fit()

  dictionary[model_name]['energy X'] = model.params[0]

  return model

def fit_latency(dictionary, model_name):
  model_data = dictionary[model_name]['model']

  X = model_data["response_length"]
  y = model_data["latency"]

  #X = sm.add_constant(X)

  model = sm.OLS(y, X).fit()

  dictionary[model_name]['latency'] = model.params[0]

  return model
```

# plotting

```python
for name in A100_subfolders:
  fit_energy(A100_files, name)
  fit_latency(A100_files, name)

for name in V100_subfolders:
  fit_energy(V100_files, name)
  fit_latency(V100_files, name)

for name in A40_subfolders:
  fit_energy(A40_files, name)
  fit_latency(A40_files, name)
```

```python
A100_df = pd.DataFrame(A100_files).T
V100_df = pd.DataFrame(V100_files).T
A40_df = pd.DataFrame(A40_files).T
```

```python
plt.scatter(A100_df['energy X'], A100_df['latency'], label='A100')
plt.scatter(A40_df['energy X'], A40_df['latency'], label='A40')
plt.scatter(V100_df['energy X'], V100_df['latency'], label='V100')

plt.legend()
plt.title("Energy use vs. latency")
plt.xlabel("Joules / token")
plt.ylabel("seconds / token")
```

## demonstrating curved residuals

```python
for name in A100_subfolders:
  current = fit_energy(A100_files, name)

  model_data = A100_files[name]['model']
  X = model_data["response_length"]

  plt.scatter(X, current.resid)
  plt.title(name)
  plt.xlabel('response length (tokens)')
  plt.ylabel('residual')
  plt.show()
```
